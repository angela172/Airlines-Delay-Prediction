```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("/Users/bristi/Desktop/Preprocessing/southwest_train.csv")
```

#### Removing unnecessary conditions (cancelled, diverted flights)

```python
df_notcancelled = df[df["Cancelled"] == False]
```

```python
df_notcancelled = df_notcancelled.drop('Cancelled', axis = 1)
```

```python
df1 = df_notcancelled[df_notcancelled["Diverted"] == False]
df1 = df1.drop('Diverted', axis = 1)
```

```python
null_dep_time_count = df1["DepTime"].isnull().sum()
print("Number of records with null DepTime:", null_dep_time_count)
```

```python
df1 = df1.drop("Airline", axis=1)
```

```python
df1 = df1.drop("FlightDate", axis=1)
```

#### One hot encoding of Origin, Destination

```python
df2 = df1
```

```python
df = pd.DataFrame(df1)

# Perform one-hot encoding on 'dest' column
one_hot_encoded = pd.get_dummies(df['Dest'], prefix='Dest')

# Concatenate one-hot encoded columns with the original DataFrame
df_encoded = pd.concat([df, one_hot_encoded], axis=1)

# Get the one-hot encoded columns related to 'dest'
dest_columns = [col for col in df_encoded.columns if col.startswith('Dest_')]

# Create a mapping dictionary for one-hot encoded columns to 'dest' values
inverse_mapping = {}
for index, row in df_encoded.iterrows():
    encoded_vals = row[dest_columns]
    dest_val = row['Dest']
    inverse_mapping[dest_val] = encoded_vals.to_dict()

# Displaying the one-hot encoded DataFrame and the inverse mapping
print("One-hot encoded DataFrame:")
print(df_encoded)

print("\nInverse Mapping of One-hot encoded columns to 'dest' values:")
print(inverse_mapping)
```
