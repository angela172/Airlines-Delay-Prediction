#### Training Set

```python
df_notcancelled = df[df["Cancelled"] == False]
```

```python
df_train = df_notcancelled[df_notcancelled["Diverted"] == False]
df_train = df_train.drop('Diverted', axis = 1)
```

```python
null_dep_time_count = df_train["DepTime"].isnull().sum()
print("Number of records with null DepTime:", null_dep_time_count)
```

```python
df_train = df_train.drop("Airline", axis=1)
```

```python
df_train.head()
```

```python
enc_nom_1 = (df_train.groupby('Origin').size()) / len(df_train)
enc_nom_1
```

```python
df_train['Origin_enc'] = df_train['Origin'].apply(lambda x : enc_nom_1[x])
df_train.head()
```

```python
enc_nom_2 = (df_train.groupby('Dest').size()) / len(df_train)
enc_nom_2
```

```python
df_train['Dest_enc'] = df_train['Dest'].apply(lambda x : enc_nom_2[x])
df_train.head()
```

#### Model Building - Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

Using PCA feature selection

```python
X = df_train[['Distance', 'DayofWeek', 'Origin_enc', 'DepTime', 'Month', 'DayofMonth', 'DepDelayMinutes']]
y = df_train[['ArrDelayMinutes']]
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
regressor = DecisionTreeRegressor(random_state = 0)  
  
# fit the regressor with X and Y data 
regressor.fit(X, y) 

y_pred = regressor.predict(X)
```

```python
mae = mean_absolute_error(y, y_pred)
print("Mean Absolute Error (MAE):", mae)

mse = mean_squared_error(y, y_pred)
print("Mean Squared Error (MSE):", mse)

r2score = r2_score(y, y_pred)
print("R2 Score:", r2score)
```

```python
X_test = df_test[['Distance', 'DayofWeek', 'Origin_enc', 'DepTime', 'Month', 'DayofMonth', 'DepDelayMinutes']]
y_test = df_test[['ArrDelayMinutes']]
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_test = sc_X.fit_transform(X_test)
y_test = sc_y.fit_transform(y_test)
```

```python
pred = regressor.predict(X_test)

mae = mean_absolute_error(y, y_pred)
print("Mean Absolute Error (MAE):", mae)
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error (MSE):", mse)
r2score = r2_score(y_test, pred)
print("R2 Score:", r2score)
```

##### Performing Grid Search CV on regression model based on rf features

```python
param_grid = { 
    'max_depth': [None, 5, 10, 15], 
    'min_samples_split': [2, 4, 8]
}
```

```python
from sklearn.model_selection import GridSearchCV

tuning_model=GridSearchCV(regressor2, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
```

```python
from datetime import datetime

def timer(start_time=None):
    if not start_time:
        start_time=datetime.now()
        return start_time
    elif start_time:
        thour,temp_sec=divmod((datetime.now()-start_time).total_seconds(),3600)
        tmin,tsec=divmod(temp_sec,60)
        print(thour,":",tmin,':',round(tsec,2))
```

```python
start_time=timer(None)
tuning_model.fit(X,y)
timer(start_time)
```

```python
tuning_model.best_params_
```

```python
y_pred = tuning_model.predict(X)

mae = mean_absolute_error(y, y_pred)
print("Mean Absolute Error (MAE):", mae)

mse = mean_squared_error(y, y_pred)
print("Mean Squared Error (MSE):", mse)

r2score = r2_score(y, y_pred)
print("R2 Score:", r2score)
```
